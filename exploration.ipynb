# Preparing train and val splits

```python
np.random.shuffle(M)
print(M)
```

```python
x = M[:, :60].astype(float)
y = M[:, 60]
```

```python
x
```

```python
def normalise_2darray(d2array):
    output_array = []
    for array in d2array:
        x = (array - np.mean(array)) / np.std(array)
        x[x < 0] *= -1
        x = (x-np.min(x))/(np.max(x)-np.min(x))
        output_array.append(x)
    return np.asarray(output_array)

x_norm = normalise_2darray(x)
x_norm
```

```python
x_norm.shape
```

```python
df_x = pd.DataFrame(x_norm)
df_x.describe()
```

```python
y
```

```python
y.shape
```

```python
def encode_classes(target):
    encoded_target = []
    encoding_dict = {'R': 0, 'M': 1}
    
    for x in target:
        encoded_target.append(encoding_dict.get(x))
            
    return np.asarray(encoded_target), encoding_dict

y_enc, encoding = encode_classes(y)
print(y_enc, '\n', encoding)
```

```python
print(x_norm.shape, y_enc.shape)
```

```python
corr_arr = []
for idx, row in enumerate(x_norm):
    arr = np.asarray(list(row) + [y_enc[idx]])
    corr_arr.append(arr)
corr_arr = np.asarray(corr_arr)
corr_map = np.corrcoef(corr_arr, rowvar=False).round(2)
corr_map = corr_map[:, 60]  # keep only final column of the heatmap | correlation to target class
corr_map = corr_map.reshape((61, 1))
```

```python
to_drop = []
for idx, value in enumerate(corr_map):
    if value > -0.1 and value < 0.1:
        to_drop.append(idx)
to_drop
```

```python
len(to_drop)
```

```python
x_norm.shape
```

```python
x_norm = np.delete(x_norm, to_drop, axis=1)
x_norm.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val = train_test_split(x_norm, test_size=0.3)
y_train, y_val = train_test_split(y_enc, test_size=0.3)
```

```python
print(x_train.shape, x_val.shape)
print(y_train.shape, y_val.shape)
```

# Creating model in Keras (tensorflow)

```python
from keras.models import Model
from keras.layers import Input, Dense
from keras.activations import sigmoid, relu, hard_sigmoid, tanh, softmax

input = Input(shape=(x_norm.shape[1], ), name="input")
dense1 = Dense(128, activation=tanh)(input)
dense2 = Dense(128, activation=sigmoid)(dense1)
dense3 = Dense(128, activation=tanh)(dense2)
dense4 = Dense(128, activation=sigmoid)(dense3)
dense5 = Dense(64, activation=tanh)(dense4)
dense6 = Dense(64, activation=sigmoid)(dense5)
dense7 = Dense(64, activation=tanh)(dense6)
# dense8 = Dense(64, activation=hard_sigmoid)(dense7)
dense_out = Dense(1, activation=hard_sigmoid, name="dense_out")(dense7)
model = Model(input, dense_out, name='test_model')
```

```python
from keras.optimizers import Adam, SGD, RMSprop
from keras.losses import BinaryCrossentropy
from keras.metrics import BinaryAccuracy, Recall

# optimizer = Adam(learning_rate=0.001, beta_1=0.8, beta_2=0.9)
optimizer = RMSprop(learning_rate=0.0025, rho=0.8)
model.compile(loss=BinaryCrossentropy(from_logits=False), optimizer=optimizer, metrics=[BinaryAccuracy(), Recall()])
```

```python
val_data = (x_val, y_val)
model.fit(x=x_train, y=y_train, validation_data=val_data, batch_size=25, epochs=1000, verbose=1)
```

```python
pred = model.predict(x=val_data[0])
pred
```

```python
model.evaluate(x=val_data[0], y=val_data[1])
```
